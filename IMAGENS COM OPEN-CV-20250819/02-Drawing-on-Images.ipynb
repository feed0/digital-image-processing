### Rectangles

* img Image.
* pt1 Vertex of the rectangle.
* pt2 Vertex of the rectangle opposite to pt1 .
* color Rectangle color or brightness (grayscale image).
* thickness Thickness of lines that make up the rectangle. Negative values, like #FILLED,mean that the function has to draw a filled rectangle.
* lineType Type of the line. See #LineTypes
* shift Number of fractional bits in the point coordinates.

```python
cv2.rectangle(B,pt1=(284,2),pt2=(410,128),color=(0,0,255),thickness=5)
cv2.rectangle(B,pt1=(284,2),pt2=(510,128),color=(0,255,0),thickness= 1)
plt.imshow(B)
```

```python
C = newblank()
cv2.rectangle(C,pt1=(284,2),pt2=(410,128),color=(255,0,0),thickness=5)
plt.imshow(C)
```

```python
# pt1 = top left
# pt2 = bottom right
cv2.rectangle(blank_img,pt1=(284,2),pt2=(510,128),color=(0,255,0),thickness=5)
```

## Circles

```python
C1 = newblank()
cv2.circle(C1, center=(260,260), radius=50, color=(255,0,0), thickness=5)
cv2.circle(C1, center=(260,260), radius=100, color=(0,255,0), thickness=5)
cv2.circle(C1, center=(260,260), radius=1, color=(0,0,255), thickness=5)
plt.imshow(C1)
```

### Filled In

```python
C2 = newblank()
cv2.circle(C2, center=(400,400), radius=50, color=(20,155,155), thickness=-1)
plt.imshow(C2)
```

### Lines

```python
# Draw a diagonal blue line with thickness of 5 px
L1 = newblank()
cv2.line(L1,pt1=(0,0),pt2=(511,511),color=(10, 55, 155),thickness=5)
plt.imshow(L1)
```

### Polygons

To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2 where ROWS are number of vertices and it should be of type int32. 


```python
blank_img1 = np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
cv2.polylines(blank_img1,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img1)
```
